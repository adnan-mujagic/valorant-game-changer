#### Feature Analysis
This is not very tidy, but I put it here for completeness and some context behind how I actually derived most of the charts used in the Feature part of the research paper.

```python
import pandas as pd
dataset = pd.read_excel("./dataset.xlsx")
wins = {}
count = {}

#Get average win for each numbers advantage element
for i in range(len(dataset["Numbers adv"])):
    cell = dataset["Numbers adv"][i]
    result = dataset["Result"][i]
    if (cell in wins.keys()):
        #Adds 1 to wins if result is win, adds 0 if result is defeat
        wins[cell] = wins[cell] + result
        #Adds one to the count of appearances
        count[cell] = count[cell] + 1
    else:
        wins[cell] = 0 + result
        count[cell] = 1
        
avg = {}
for i in wins.keys():
    avg[i] = wins[i] / count[i]
    
avg

```

```python
import matplotlib.pyplot as plt

plt.plot(avg.keys(), avg.values())
plt.xlabel('Numbers advantage')
plt.ylabel('Win percentage')
plt.show()
```

```python
def cluster(data, results, tolerance):
    wins = {}
    count = {}
    for i in range(len(data)):
        cell = data[i]
        result = results[i]
        key = cell//tolerance * tolerance
        if (key in wins.keys()):
            wins[key] += result
            count[key] += 1
        else:
            wins[key] = 0 + result
            count[key] = 1
    return dict(sorted(wins.items())), dict(sorted(count.items()))

wins, count = cluster(dataset["Health advantage"], dataset["Result"], 300)


```

```python
import matplotlib.pyplot as plt

avg = {}

for i in count.keys():
    avg[i] = wins[i] / count[i]

plt.plot(avg.keys(), avg.values())
plt.xlabel('Health advantage')
plt.ylabel('Win percentage')
plt.show()
```
